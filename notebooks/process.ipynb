```python
%%sql

create or replace table opensea_events as (
    select
        *
    from
        read_json_auto('../messages/*.jsonl')
);
select
    *
from
    opensea_messages;
```

```python
%%sql

create or replace view collection_events as (
    select
        *
    from
        opensea_events
    where
        topic = 'collection:*'
        and event != 'phx_reply'
);
select
    *
from
    collection_events;
```
